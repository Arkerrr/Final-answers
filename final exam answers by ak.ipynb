A construction site keeps daily records of concrete truck deliveries. Each record contains the truck ID and the delivered volume (in cubic meters).

You are given the following list of delivery records:

deliveries = [
    ("T01", 5.5),
    ("T02", 7.0),
    ("T01", 4.5),
    ("T03", 6.0),
    ("T02", 5.0),
]

**(a) Write a function to compute total volumes — 12 marks**

Create a function named total_volume(deliveries).

The function must:


*   take the list of deliveries as an input parameter.
*   create an empty dictionary to store total volumes for each truck.
*   loop through each tuple in the list.
*   add the delivered volume to the correct truck ID inside the dictionary.


The function must return a dictionary in the following format:

{

    "T01": total volume delivered by T01,

    "T02": total volume delivered by T02,

    "T03": total volume delivered by T03
}

Example of expected output (your program must calculate it):
{
    "T01": 10.0,

    "T02": 12.0,

    "T03": 6.0
}


```python
deliveries = [
    ("T01", 5.5),
    ("T02", 7.0),
    ("T01", 4.5),
    ("T03", 6.0),
    ("T02", 5.0),
]

def total_volume(deliveries):
    totals = {}
    for truck_id, volume in deliveries:
        if truck_id in totals:
            totals[truck_id]+= volume
        else:
            totals[truck_id]=volume
    return totals  

print(total_volume(deliveries))  
    
```


**(b) Find the truck with the highest total volume — 8 marks**

After calling your total_volume() function:

Write code (outside the function) to find which truck delivered the largest amount of concrete in total.


*   Store the truck ID and its total volume.
*   Print the result using this exact format:


Top truck: T02 delivered 12.0 m3 (Your numbers should match the actual computed values.)

```python

totals = total_volume(deliveries)
Top_truck = None
Top_volume = 0
for truck_id, volume in totals.items():
    if volume >Top_volume:
        Top_truck= truck_id
        Top_volume= volume

print(f"Top truck: {Top_truck} delivered {Top_volume} m3")
```

**TASK 1 - 8 marks**

```python

water_levels = [2.1, 2.3, 2.8, 3.0, 3.5, 4.2, 4.8, 5.1, 4.9, 4.0, 3.7, 3.2, 2.9, 2.4]
print(f"Day 1 water level: {water_levels[0]}")
print(f"Last Day water level: {water_levels[-1]}")

print("Water levels for Day 5 to 8,")
for i in range(4,8):
    print(f"Day {i+1} : {water_levels[i]}")


max_level = water_levels[0]
for i in range(len(water_levels)):
    if water_levels[i] > max_level:
        max_level = water_levels[i]
        max_day = i+1
print(f"Maximum water level: {max_level} m on Day {max_day}")
```

**TASK 2 - 8 marks**

Write a loop that goes through all 14 days and prints the risk category for each day using the rules:

If level < 2.5 → "Too Low"

If 2.5 ≤ level ≤ 4.5 → "Normal"

If level > 4.5 → "High"


Your output format should be:

Day 1: 2.1 m → Too Low

Day 2: 2.3 m → Too Low

Day 3: 2.8 m → Normal

```python
water_levels = [2.1, 2.3, 2.8, 3.0, 3.5, 4.2, 4.8, 5.1, 4.9, 4.0, 3.7, 3.2, 2.9, 2.4]

for i in range(len(water_levels)):
    level = water_levels[i]

    if level<2.5:
        category = "Too Low"
    elif 2.5 <= level <= 4.5:
        category = "Normal"
    else:
        category =  "High"
    print(f"Day{i+1}: {level}m --> {category}")
```
